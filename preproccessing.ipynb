```python
import pandas as pd
import os
```

```python
image_folder = "images/images_normalized"  # Update this if images are in a different location
reports_path = "indiana_reports.csv"
projections_path = "indiana_projections.csv"
output_csv = "processed_xray_data.csv"
```

```python
df_reports = pd.read_csv(reports_path)
df_projections = pd.read_csv(projections_path)
```

```python
df_merged = df_projections.merge(df_reports, on="uid", how="inner")
```

```python
df_merged = df_merged[["filename", "projection", "findings", "impression"]]
```

```python
df_cleaned = df_merged.dropna(subset=["findings", "impression"])
```

```python
df_cleaned.to_csv(output_csv, index=False)
```

```python
print(f"Processed dataset saved to {output_csv}, with {len(df_cleaned)} valid samples.")
```
